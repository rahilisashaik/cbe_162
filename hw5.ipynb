```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step
```

# Problem 3A

```python
numerator = [3]
denominator = [5, 1]
system = TransferFunction(numerator, denominator) # should perform inverse laplace transform back to time domain numerically

t = np.linspace(0, 30, 1000)
t, y = step(system, T=t)
y = 1.5 * y
```

```python
print(system)
```

```python
plt.plot(t, y)
plt.title("Y(s)/U(s) with Step Input of 1.5")
plt.xlabel("Time (s)")
plt.ylabel("Response y(t)")
plt.grid(True)
plt.show()
```

# Problem 3B

```python
res = [0] * 20
for i in range(21):
    res[-i] = y[-i]

print(res)
```

The process gain is 4.5 seconds as this is the asymptotic value that y(t) approaches
