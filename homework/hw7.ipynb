```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp
```

# Problem 1A

```python
numerator = [0.9, 0.153]
denominator = [1, 0.8, 0.24]

system = sp.TransferFunction(numerator, denominator)
zeroes, poles, _ = sp.tf2zpk(numerator, denominator)
print("poles:", poles)
print("zeroes:", zeroes)
```

# Problem 1B (First-order Pade Approximation)

```python
# first order approximation
pade_numerator = [1, -1]
pade_denominator = [1, 1]

numerator_sys = [0.9, 0.153]
denominator_sys = [1, 0.8, 0.24]

numerator = np.polymul(pade_numerator, numerator_sys)
denominator = np.polymul(pade_denominator, denominator_sys)

zeroes, poles, _ = sp.tf2zpk(numerator, denominator)

for i, zero in enumerate(zeroes):
    print(f"zero {i+1}: ", zero)

for i, pole in enumerate(poles):
    print(f"pole {i+1}: ", pole)
```

```python
plt.figure(figsize=(20,6))
plt.scatter(np.real(zeroes), np.imag(zeroes), label="zeroes")
plt.scatter(np.real(poles), np.imag(poles), label="poles")
plt.xlabel("real")
plt.ylabel("imaginary")
plt.title("poles and zeroes of first order pade approx of G_delay(s)")
plt.grid(True)
plt.legend()
```

# Problem 1B (Second-order Pade Approximation)

```python
# second order approximation
pade_numerator = [1, -1, 1]
pade_denominator = [1, 1, -1]

numerator_sys = [0.9, 0.153]
denominator_sys = [1, 0.8, 0.24]

numerator = np.polymul(pade_numerator, numerator_sys)
denominator = np.polymul(pade_denominator, denominator_sys)

zeroes, poles, _ = sp.tf2zpk(numerator, denominator)

for i, zero in enumerate(zeroes):
    print(f"zero {i+1}: ", zero)

for i, pole in enumerate(poles):
    print(f"pole {i+1}: ", pole)
```

```python
plt.figure(figsize=(20,6))
plt.scatter(np.real(zeroes), np.imag(zeroes), label="zeroes")
plt.scatter(np.real(poles), np.imag(poles), label="poles")
plt.xlabel("real")
plt.ylabel("imaginary")
plt.title("poles and zeroes of second order pade approx of G_delay(s)")
plt.grid(True)
plt.legend()
```

# Problem 3C

```python
def compute_roots(Kc):
    coefficients = [1, 2, 5 + Kc]
    roots = np.roots(coefficients)
    return roots

Kc_values = [-5, 0, -3]

for Kc in Kc_values:
    roots = compute_roots(Kc)
    for root in roots:
        print(f"For Kc = {Kc}, the zeroes are: {root}")
```
