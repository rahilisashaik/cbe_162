```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step
```

### Plotting transfer function without time delay

```python
tau_C = 5

numerator = [8, 6, 1]
denominator = [tau_C, 3*tau_C, tau_C - 4]

G_s = TransferFunction(numerator, denominator)
t = np.linspace(0, 50, 500)
time, resp = step(G_s, T=t)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time, resp)
plt.xlabel('Time (s)')
plt.ylabel('Response')
plt.title('Step Response of the System with no time delay')
plt.grid()
plt.legend()
plt.show()
```

### Now, let's plot the transfer function with the time delay Pade approximated 

```python
tau_C = 5

numerator = [4, 11, 6.5, 1]
denominator = [tau_C, 4*tau_C, 6*tau_C, 4*tau_C - 4]

G_s_td = TransferFunction(numerator, denominator)
t = np.linspace(0, 50, 500)
time, resp = step(G_s_td, T=t)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time, resp)
plt.xlabel('Time (s)')
plt.ylabel('Response')
plt.title('Step Response of the System with pade approximated time delay')
plt.grid()
plt.legend()
plt.show()
```
