```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step
```

```python
time = np.linspace(0, 25, 1000)

num_i = [2.25, 3.375] 
den_i = [8, 17, 10, 0]
tf_i = TransferFunction(num_i, den_i)

num_ii = [2.25, 3.375]
den_ii = [16, 24, 10, 0] 
tf_ii = TransferFunction(num_ii, den_ii)
num_iii = [2.25, 3.375]  
den_iii = [8, 11, 2.5, 0]
tf_iii = TransferFunction(num_iii, den_iii)

t_i, y_i = step(tf_i, T=time)
t_ii, y_ii = step(tf_ii, T=time)
t_iii, y_iii = step(tf_iii, T=time)
```

```python

plt.figure(figsize=(20, 6))
plt.plot(t_i, y_i, label='Case (i): Original Transfer Functions')
plt.plot(t_ii, y_ii, label='Case (ii): Modified Gp')
plt.plot(t_iii, y_iii, label='Case (iii): Modified Gd')

plt.title("Step Response of Y for a Step Change in D at t=5")
plt.xlabel("Time (t)")
plt.ylabel("Response (Y)")
plt.legend()
plt.grid()

plt.show()
```
